```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from google.colab import files

# Fazer upload do CSV manualmente
uploaded = files.upload()

# Pegar nome do arquivo enviado
arquivo = list(uploaded.keys())[0]

# Ler CSV
df = pd.read_csv(arquivo)

# Criar gráfico
media = df.groupby('Destino')['Tempo Total (min)'].mean()

plt.figure(figsize=(8,5))
media.plot(kind='bar')

plt.title('Tempo Médio por Destino')
plt.xlabel('Destino')
plt.ylabel('Tempo Médio (min)')
plt.xticks(rotation=0)

# Salvar gráfico como imagem
plt.savefig('grafico.png')
plt.close()

# Criar PDF
pdf = PdfPages('relatorio.pdf')

# Página do gráfico
fig, ax = plt.subplots(figsize=(10,6))
img = plt.imread('grafico.png')

ax.imshow(img)
ax.axis('off')

pdf.savefig(fig, bbox_inches='tight')
pdf.close()

print("PDF criado com sucesso!")

# Baixar PDF
files.download('relatorio.pdf')
```

```python
media_transito = df.groupby('Nível de Trânsito')['Tempo Total (min)'].mean()

plt.figure(figsize=(8,5))
media_transito.plot(kind='bar')

plt.title('Tempo Médio por Nível de Trânsito')
plt.xlabel('Nível de Trânsito')
plt.ylabel('Tempo Médio (min)')
plt.xticks(rotation=0)

plt.show()
```
